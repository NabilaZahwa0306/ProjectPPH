## Data loading

### Subtask:
Load the data from "bquxjob_66265804_1972b72867c.csv" into a dataframe.


**Reasoning**:
Import pandas and load the CSV file into a DataFrame, then display the head to verify the data loading.



```python
import pandas as pd

df = pd.read_csv('bquxjob_66265804_1972b72867c.csv')
display(df.head())
```

## Data exploration

### Subtask:
Explore the data to understand its structure and content, identifying columns suitable for analysis and visualization.


**Reasoning**:
Display the column names and their data types, show the first few rows, get descriptive statistics for numerical columns, and check for missing values to explore the data structure and content.



```python
display(df.info())
display(df.head())
display(df.describe())
display(df.isnull().sum())
```

## Data analysis

### Subtask:
Perform data analysis relevant to generating bar and line plots. This might involve grouping data, calculating statistics, or preparing data for plotting based on the content of the CSV.


**Reasoning**:
The subtask requires grouping, calculating means, and creating pivot tables, which can be achieved in a single code block.



```python
df_grouped = df.groupby(['skenario', 'tahun']).mean().reset_index()
display(df_grouped)

df_pivot_pendapatan = df.pivot_table(index='tahun', columns='skenario', values='pendapatan')
display(df_pivot_pendapatan)

df_pivot_pph_badan = df.pivot_table(index='tahun', columns='skenario', values='pph_badan')
display(df_pivot_pph_badan)
```

## Data visualization

### Subtask:
Create colored bar and line plots based on the analysis, ensuring they are comprehensive and valid representations of the data.


**Reasoning**:
Generate the specified line and bar plots using the prepared dataframes and display them with appropriate titles and labels.



```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# Line plot for 'pendapatan' over 'tahun' for each 'skenario'
sns.lineplot(data=df_pivot_pendapatan, markers=True, ax=axes[0])
axes[0].set_title('Pendapatan over Tahun by Skenario')
axes[0].set_xlabel('Tahun')
axes[0].set_ylabel('Pendapatan')

# Bar plot for 'pph_badan' for each 'skenario' and 'tahun'
sns.barplot(data=df_grouped, x='tahun', y='pph_badan', hue='skenario', ax=axes[1])
axes[1].set_title('PPH Badan by Skenario and Tahun')
axes[1].set_xlabel('Tahun')
axes[1].set_ylabel('PPH Badan')

plt.tight_layout()
plt.show()
```
