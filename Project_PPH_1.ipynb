## Data loading

### Subtask:
Load the data from "Fixed_Assets.csv", "Financial_Data_2020_2025.csv", and "Fiscal_Policy.csv" into dataframes.


**Reasoning**:
Import pandas and load the three CSV files into dataframes as instructed.



```python
import pandas as pd

df_fixed_assets = pd.read_csv('Fixed_Assets.csv')
df_financial_data = pd.read_csv('Financial_Data_2020_2025.csv')
df_fiscal_policy = pd.read_csv('Fiscal_Policy.csv')

display(df_fixed_assets.head())
display(df_financial_data.head())
display(df_fiscal_policy.head())
```

## Data exploration

### Subtask:
Explore the loaded dataframes to understand their structure, columns, and data types.


**Reasoning**:
Explore the loaded dataframes by printing their info, shape, descriptive statistics, and checking for missing values as requested in the instructions.



```python
print("--- df_fixed_assets Information ---")
df_fixed_assets.info()
print("\n--- df_fixed_assets Shape ---")
print(df_fixed_assets.shape)
print("\n--- df_fixed_assets Descriptive Statistics ---")
display(df_fixed_assets.describe())
print("\n--- df_fixed_assets Missing Values ---")
print(df_fixed_assets.isnull().sum())

print("\n--- df_financial_data Information ---")
df_financial_data.info()
print("\n--- df_financial_data Shape ---")
print(df_financial_data.shape)
print("\n--- df_financial_data Descriptive Statistics ---")
display(df_financial_data.describe())
print("\n--- df_financial_data Missing Values ---")
print(df_financial_data.isnull().sum())

print("\n--- df_fiscal_policy Information ---")
df_fiscal_policy.info()
print("\n--- df_fiscal_policy Shape ---")
print(df_fiscal_policy.shape)
print("\n--- df_fiscal_policy Descriptive Statistics ---")
display(df_fiscal_policy.describe())
print("\n--- df_fiscal_policy Missing Values ---")
print(df_fiscal_policy.isnull().sum())
```

## Data cleaning

### Subtask:
Clean the dataframes by handling missing values, outliers, or any inconsistencies identified during exploration.


**Reasoning**:
Confirm that there are no missing values in the dataframes as instructed.



```python
print("Missing values in df_fixed_assets:")
print(df_fixed_assets.isnull().sum())

print("\nMissing values in df_financial_data:")
print(df_financial_data.isnull().sum())

print("\nMissing values in df_fiscal_policy:")
print(df_fiscal_policy.isnull().sum())
```

## Data wrangling

### Subtask:
Merge or combine the dataframes as needed for analysis, based on common columns or relevant information.


**Reasoning**:
Merge the `df_financial_data` and `df_fiscal_policy` dataframes on the 'tahun' column and display the head of the merged dataframe.



```python
df_merged = pd.merge(df_financial_data, df_fiscal_policy, on='tahun')
display(df_merged.head())
```

## Data analysis

### Subtask:
Perform data analysis to derive insights from the combined data, focusing on trends, relationships, or patterns related to financial data and fiscal policy.


**Reasoning**:
Calculate Net Income Before Tax, the tax amount considering the tax holiday, and Net Income After Tax. Then display the updated dataframe.



```python
df_merged['Net_Income_Before_Tax'] = df_merged['pendapatan'] - df_merged['beban_operasional'] - df_merged['penyusutan']

def calculate_tax(row):
    if row['tahun'] >= row['tax_holiday_awal'] and row['tahun'] <= row['tax_holiday_akhir']:
        return 0
    else:
        return row['Net_Income_Before_Tax'] * row['tax_rate']

df_merged['Tax_Amount'] = df_merged.apply(calculate_tax, axis=1)
df_merged['Net_Income_After_Tax'] = df_merged['Net_Income_Before_Tax'] - df_merged['Tax_Amount']

display(df_merged)
```

**Reasoning**:
Analyze the trends of revenue, operating expenses, depreciation, and net income (before and after tax) over the years and the impact of the tax holiday by plotting these trends.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.lineplot(x='tahun', y='pendapatan', data=df_merged, marker='o', label='Revenue', color='skyblue')
sns.lineplot(x='tahun', y='beban_operasional', data=df_merged, marker='o', label='Operating Expenses', color='salmon')
sns.lineplot(x='tahun', y='penyusutan', data=df_merged, marker='o', label='Depreciation', color='lightgreen')
sns.lineplot(x='tahun', y='Net_Income_Before_Tax', data=df_merged, marker='o', label='Net Income Before Tax', color='gold')
sns.lineplot(x='tahun', y='Net_Income_After_Tax', data=df_merged, marker='o', label='Net Income After Tax', color='purple')

plt.title('Financial Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.grid(True)
plt.legend()
plt.xticks(df_merged['tahun'])
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x='tahun', y='Net_Income_After_Tax', data=df_merged, marker='o', color='purple')
plt.title('Net Income After Tax Trend and Tax Holiday Impact')
plt.xlabel('Year')
plt.ylabel('Net Income After Tax')
plt.axvspan(df_merged['tax_holiday_awal'].iloc[0], df_merged['tax_holiday_akhir'].iloc[0], color='gray', alpha=0.3, label='Tax Holiday Period')
plt.grid(True)
plt.legend()
plt.xticks(df_merged['tahun'])
plt.tight_layout()
plt.show()
```

## Data visualization

### Subtask:
Create comprehensive and visually appealing charts and graphs to represent the findings from the data analysis. Ensure the visualizations are colorful and valid.


**Reasoning**:
Generate visualizations based on the analyzed data to fulfill the subtask requirements. This includes creating a bar chart comparing net income before and after tax, a stacked bar chart of revenue breakdown, and a line chart of the tax rate trend.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Bar chart comparing Net_Income_Before_Tax and Net_Income_After_Tax
plt.figure(figsize=(12, 7))
bar_width = 0.35
x = df_merged['tahun']
plt.bar(x - bar_width/2, df_merged['Net_Income_Before_Tax'], bar_width, label='Net Income Before Tax', color='skyblue')
plt.bar(x + bar_width/2, df_merged['Net_Income_After_Tax'], bar_width, label='Net Income After Tax', color='salmon')
plt.title('Comparison of Net Income Before and After Tax by Year')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.xticks(x)
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()

# 2. Stacked bar chart showing revenue breakdown
plt.figure(figsize=(12, 7))
plt.bar(df_merged['tahun'], df_merged['beban_operasional'], label='Operating Expenses', color='lightcoral')
plt.bar(df_merged['tahun'], df_merged['penyusutan'], bottom=df_merged['beban_operasional'], label='Depreciation', color='gold')
plt.bar(df_merged['tahun'], df_merged['Tax_Amount'], bottom=df_merged['beban_operasional'] + df_merged['penyusutan'], label='Tax Amount', color='lightgreen')
plt.bar(df_merged['tahun'], df_merged['Net_Income_After_Tax'], bottom=df_merged['beban_operasional'] + df_merged['penyusutan'] + df_merged['Tax_Amount'], label='Net Income After Tax', color='skyblue')
plt.title('Revenue Breakdown by Component and Year')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.xticks(df_merged['tahun'])
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()

# 3. Line chart showing the trend of the tax rate
plt.figure(figsize=(10, 6))
sns.lineplot(x='tahun', y='tax_rate', data=df_fiscal_policy, marker='o', color='purple')
plt.title('Tax Rate Trend Over the Years')
plt.xlabel('Year')
plt.ylabel('Tax Rate')
plt.xticks(df_fiscal_policy['tahun'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

*   Net Income After Tax shows a significant increase during the tax holiday period (2021-2022) compared to the years before and after, indicating the positive financial impact of the policy.

*   Operating Expenses (`beban_operasional`) also increased over the years, but at a slower pace than revenue.
*   Depreciation (`penyusutan`) remained relatively stable or increased slightly across the years.
